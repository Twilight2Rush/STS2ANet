```python
# get a wohle years data in manhattan
data_path_list = [
  '../data/raw data/yellow_tripdata_2022-01.parquet',
  # '../data/raw data/yellow_tripdata_2022-02.parquet',
  # '../data/raw data/yellow_tripdata_2022-03.parquet',
  # '../data/raw data/yellow_tripdata_2022-04.parquet',
  # '../data/raw data/yellow_tripdata_2022-05.parquet',
  # '../data/raw data/yellow_tripdata_2022-06.parquet',
  # '../data/raw data/yellow_tripdata_2022-07.parquet',
  # '../data/raw data/yellow_tripdata_2022-08.parquet',
  # '../data/raw data/yellow_tripdata_2022-09.parquet',
  # '../data/raw data/yellow_tripdata_2022-10.parquet',
  # '../data/raw data/yellow_tripdata_2022-11.parquet',
  # '../data/raw data/yellow_tripdata_2022-12.parquet'
]
start_time = pd.to_datetime('2022/01/01')
end_time = pd.to_datetime('2022/01/14')
m_tripdata = dp.get_tripdata(data_path_list, id_list, start_time, end_time)
m_tripdata
```

```python
def od_construct(tripdata,start_time, end_time):
  hour_count = tripdata.groupby(['PULocationID','DOLocationID','date','day','hour']).agg({'count':'sum'})
  hour_count = hour_count.reset_index(drop=False)
  grouped = hour_count.groupby(['date', 'hour'])
  temp_adj_dict = {}
  adj_dict = {}
  for group_name, group_data in grouped:
    adjacency_matrix = np.zeros((len(id_list), len(id_list)))
    id_mapping = {location: idx for idx, location in enumerate(id_list.values)}
    for _, row in group_data.iterrows():
        pu_location = id_mapping[row['PULocationID']]
        do_location = id_mapping[row['DOLocationID']]
        count = row['count']
        adjacency_matrix[pu_location][do_location] = count
    temp_adj_dict[group_name] = adjacency_matrix
  date_range = pd.date_range(start_time,end_time)
  key_to_check_list = []
  for date in date_range:
    for h in range(0,24):
      key_to_check_list.append((date.date(),h))
  for k in key_to_check_list:
    if k not in temp_adj_dict.keys():
      adj_dict[k] = np.zeros((len(id_list), len(id_list)))
    else:
      adj_dict[k] = temp_adj_dict[k]
  return adj_dict
    
  
```

```python
od_adj = od_construct(m_tripdata,'2022-01-01','2022-01-14')
```

```python
def is_cross_day_hour_tail(hour):
  if hour >= 0 and hour <=6:
    return True
  else:
    return False
    
def is_cross_day_hour_head(hour):
  if hour >= 18 and hour <=23:
    return True
  else:
    return False
```

```python
def get_cross_day_emb(t): 
  weekends = [6,7]
  cd1 = cd2 = cd3 = 0
  if is_cross_day_hour_head(t[1]):
    cd3 = 1
    if t[0].weekday()+1 in weekends:
      cd1 = 1
      if t[0].weekday()+1 == 6:
        cd2 = 1
  elif is_cross_day_hour_tail(t[1]):
    cd3 = 1
    if t[0].weekday()+1 in weekends:
      cd2 = 1
      if t[0].weekday()+1 == 7:
        cd1 = 1
  return np.array([cd1,cd2,cd3])
```

```python
def t_emb(t):
  d_w_list = range(1,8) # day of  week
  d_w_oh_list = np.eye(len(d_w_list)) # one hot
  d_w_emb = d_w_oh_list[d_w_list.index(t[0].weekday()+1)] # day of week embedding
  h_d_list = range(0,24) # hour of day
  h_d_oh_list = np.eye(len(h_d_list)) # one hot
  h_d_emb = h_d_oh_list[h_d_list.index(t[1])]
  cd_emb = get_cross_day_emb(t)
  emb = np.hstack((d_w_emb,h_d_emb,cd_emb))
  return emb
```
