```python
import argparse
import utils
import pandas as pd
```

```python
def parse_args():
    """
    Description: args
    """
    parser = argparse.ArgumentParser()

    parser.add_argument(
        "--raw_zone_json_path", nargs="?", default="../data/raw data/NYC Taxi Zones.geojson", help="taxi zone geojson path"
    )
    
    parser.add_argument(
        "--zone_json_path", nargs="?", default="../data/processed data/zones.geojson", help="processed taxi zone geojson path"
    )
    
    parser.add_argument(
        "--cd_adj_path", nargs="?", default="../data/processed data/cd_adj.csv", help="the centroid distance adjacency matrix path"
    )
    
    parser.add_argument(
        "--d_adj_path", nargs="?", default="../data/processed data/d_adj.csv", help="the space distance adjacency matrix path"
    )
    
    parser.add_argument(
        "--adj_path", nargs="?", default="../data/processed data/adj.csv", help="the adjacency matrix path"
    )
    
    parser.add_argument(
        "--w_adj_path", nargs="?", default="../data/processed data/w_adj.csv", help="the weighted adjacency matrix path"
    )
    
    parser.add_argument(
        "--q", nargs="?", default=0.1, help="the quantile number of distance"
    )
    
    parser.add_argument(
        "--raw_poi_json_path", nargs="?", default="../data/raw data/Points Of Interest.geojson", help="raw poi json path"
    )
    
    parser.add_argument(
        "--poi_json_path", nargs="?", default="../data/processed data/poi.geojson", help="processed poi json path"
    )
    
    parser.add_argument(
        "--poi_feature_path", nargs="?", default="../data/processed data/poi_f.csv", help="the poi feature path"
    )

    return parser.parse_known_args()[0]
```

```python
args = parse_args()
```
