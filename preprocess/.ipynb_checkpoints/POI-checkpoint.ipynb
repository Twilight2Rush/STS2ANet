### Import Packages

```python
import pandas as pd
import numpy as np
import data_preprocess as dp
import data_visualization as dv
import geopandas as gpd
from shapely.geometry import Point, LineString
import matplotlib.pyplot as plt
```

### Read Data

```python
def get_poi(json_path):
  poi_gdf = gpd.read_file(json_path)
  m_poi = poi_gdf['borough'].apply(lambda x:x=='1')
  m_poi_gdf = poi_gdf[m_poi]
  return m_poi_gdf
```

```python
json_path = './data/Points Of Interest.geojson'
poi_gdf = get_poi(json_path)
poi_gdf
```

```python
def get_zones(json_path):
  gdf = gpd.read_file(json_path) # read taxi zone file
  m_zones = gdf['borough'].apply(lambda x:x=='Manhattan')
  m_remove103 = gdf['location_id'].apply(lambda x:x!='103')
  m_gdf = gdf[(m_zones) & (m_remove103)]
  m_gdf.set_index('objectid',inplace=True)
  return m_gdf
```

```python
zone_json_path = './data/NYC Taxi Zones.geojson'
zone_gdf = get_zones(zone_json_path)
zone_gdf
```

```python
def getZone(point,zone_gdf):
  for index,zone in zone_gdf.iterrows():
    if point.within(zone['geometry']):
      return zone['location_id']
  return 1000 # not in any zone
    
```

```python
poi_gdf['zone_id'] = poi_gdf['geometry'].apply(getZone, zone_gdf=zone_gdf)
```

```python
def poi_vec(loc_id,poi_gdf):
  poi_list = np.zeros(13)
  loc_gdf = poi_gdf[poi_gdf['zone_id'].apply(lambda x:x==loc_id)]
  value_counts = loc_gdf['facility_t'].value_counts(normalize=True)
  for i, v in value_counts.items():
    poi_list[int(i)-1] = v
  return poi_list
  
```

```python
zone_gdf['poi_vec'] = zone_gdf['location_id'].apply(poi_vec, poi_gdf=poi_gdf)
zone_gdf['poi_vec']
```

```python
from sklearn.cluster import KMeans
iteration = 500  # 聚类最大迭代次数
model = KMeans(n_clusters=5, max_iter=iteration) 
poi_vec = np.stack(zone_gdf['poi_vec'].values)
model.fit(poi_vec)  # 开始聚类
labels = model.labels_
zone_gdf['label'] = labels
print('labels: \n', model.labels_)
print('cluster_centers_: \n', model.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
id_list = zone_gdf.index.to_numpy()
plt.figure(figsize=(32,6))
colormap = ['red','cyan','yellow','green','blue']
color_list = []
for v in zone_gdf['label'].values:
  color_list.append(colormap[v-1])
plt.bar(range(0,len(zone_gdf)), zone_gdf['label'],tick_label=id_list,color=color_list)
plt.show()
```

```python
# 获取每个数据点的聚类标签
predicted_labels = model.labels_

# 使用PCA进行数据降维
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(poi_vec)

# 绘制K-means预测的聚类结果
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=predicted_labels, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering (PCA)')
plt.show()
```

```python
# 获取每个数据点的聚类标签
predicted_labels = model.labels_

# 使用t-SNE进行数据降维
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
reduced_data = tsne.fit_transform(poi_vec)

# 绘制K-means预测的聚类结果
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=predicted_labels, cmap='viridis')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('K-means Clustering (t-SNE)')
plt.show()
```
